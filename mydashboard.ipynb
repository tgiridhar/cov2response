```python
import pandas as pd 
import matplotlib.ticker as tkr
import datetime
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings('ignore')

print(f'Generated at {datetime.datetime.now()}')
g=!ls ../csse_covid_19_daily_reports/*.csv
al=[]
for f in g:
    if pd.to_datetime(f.split('/')[-1][:10]) > pd.to_datetime('2020-04-01'):      
        d=pd.read_csv(f)
        d['dt']=pd.to_datetime(f.split('/')[-1][:10])
        al.append(d)
adt=pd.concat(al)
from IPython.core.display import HTML,display,Image
display(HTML('<a href="http://www.cov2response.com"> Back to cov2response.com </a>'))
```
